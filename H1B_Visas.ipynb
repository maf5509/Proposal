```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
visa = pd.read_csv('C://Users//fergu//h1b_kaggle.csv')
```

```python
states_dict = {'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA', 'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'DC': 'DC', 'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD', 'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC', 'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'}
```

```python
states_dict['Na'] = 'Na'
```

```python
visa['state'] = visa['state'].str.replace('District Of Columbia', 'DC')
```

```python
states_dict['Puerto Rico'] = 'PR'
```

```python
visa['state'].unique()
```

```python
visa['abbrev'] = [states_dict[name] for name in visa['state']]
```

```python
visa.head()
```

```python
print(plt.style.available)
```

```python
plt.style.use(['ggplot'])
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x="abbrev", hue="full_time", data=visa)
plt.xlabel('State')
plt.legend(['Part Time', 'Full Time'])
plt.show()
```

```python
visa['wage'].describe()
```

```python
visa.sort_values('wage', ascending=False).head(50)
```

```python
#plt.subplot(2,1,1)
sns.stripplot(x="year", y="wage", data=visa, jitter=True)
plt.ylim([0,10000000])
plt.ylabel('Adjusted Wage')
plt.title('Strip Plot: Salary Data for US H1-B Visa Applications, All Years')
plt.xticks(rotation=90)
plt.show()

#plt.subplot(2,1,2)
sns.stripplot(x="year", y="wage", data=visa, jitter=True)
plt.ylim([0,500000000])
plt.ylabel('Adjusted Wage')
plt.title('Strip Plot: Salary Data for US H1-B Visa Applications, All Years')
plt.xticks(rotation=90)
plt.show()
```

```python
above = [1 if amount > 200000 else 0 for amount in visa['wage']]
sum(above) / len(visa['wage'])
```

```python
visa['wage_adj'] = [value if value < 200000 else 0 for value in visa['wage']]
```

```python
#plt.subplot(1,2,1)
#sns.boxplot(x="year", y="wage", hue='full_time', data=visa)
#plt.xticks(rotation=90)
#plt.show()


#plt.subplot(1,2,2)
sns.boxplot(x="year", y="wage_adj", hue='full_time', data=visa)
plt.ylabel('Adjusted Wage')
plt.title('Boxplot: Salary Data for US H1-B Visa Applications, All Years')
#plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x="year", y="wage_adj", hue='full_time', data=visa)
plt.ylabel('Adjusted Wage')
plt.title('Boxplot: Salary Data for US H1-B Visa Applications, All Years')
plt.show()
```

```python
'''plt.subplot(1,2,1)
sns.distplot(visa[visa['full_time'] == 'N']['wage_adj'], bins=20)
plt.xlabel('Prevailing Wage')
plt.ylabel('Frequency')
plt.title('Salary Data for US H1-B Part-Time Visa Applications, All Years')'''

#plt.subplot(1,2,2)
sns.distplot(visa[visa['full_time'] == 'Y']['wage_adj'], bins=20, color='blue')
plt.xlabel('Prevailing Wage')
plt.ylabel('Frequency')
plt.title('Salary Data for US H1-B Full-Time Visa Applications, All Years')

plt.show()
```
